## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd 
import tensorflow as tf
import matplotlib.pyplot as plt

#  Import and read the charity_data.csv.
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Look at shape of dataset
application_df.shape
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=["EIN", "NAME"])
```

```python
application_df.columns
```

```python
# Determine the number of unique values in each column.
pd.Series({col:application_df[col].nunique() for col in application_df})
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_counts = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = list(application_counts[application_counts < 528].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classifications_greater_1 = application_df['CLASSIFICATION'].value_counts(ascending=False).loc[lambda x : x>1]
classifications_greater_1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classification_counts = application_df['CLASSIFICATION'].value_counts()
classifications_to_replace = list(classification_counts[classification_counts < 1883].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Generate our categorical variable lists
categorical_list = application_df.dtypes[application_df.dtypes == "object"].index.tolist()
categorical_list
```

```python
# Check the number of unique values in each column
application_df[categorical_list].nunique()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
categorical_encoded = pd.get_dummies(application_df[categorical_list])
categorical_encoded.head()
```

```python
# Concatenate the `categorical_data` encoded dummies with the DataFrame
application_df = pd.concat([application_df, categorical_encoded], axis=1)

# Display the combined DataFrame.
application_df.head()
```

```python
# Dropping categorical columns
application_df = application_df.drop(columns = categorical_list, axis=1)

application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y= application_df['IS_SUCCESSFUL'].values
X= application_df.drop(columns=['IS_SUCCESSFUL']).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled[:1]
```

```python
X_test_scaled[:1]
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  80
hidden_nodes_layer2 = 30

funding_model = tf.keras.models.Sequential()

# First hidden layer
# Add our first Dense layer, including the input layer
funding_model.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
# nn.add(tf.keras.layers.Dropout(rate=0.1, seed=2))
funding_model.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
# Add the output layer that uses a probability activation function
funding_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
funding_model.summary()
```

```python
# Compile the model
funding_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_funding_model = funding_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Create a DataFrame containing training history
funding_model_history_df = pd.DataFrame(fit_funding_model.history)

# Increase the index by 1 to match the number of epochs
funding_model_history_df.index += 1

# Plot the loss
funding_model_history_df.plot(y="loss")
plt.show()
```

```python
# Plot the accuracy
funding_model_history_df.plot(y="accuracy")
plt.show()
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = funding_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
funding_model.save('AlphabetSoupCharity.h5')
```
